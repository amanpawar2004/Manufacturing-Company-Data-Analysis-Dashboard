```python
df["Type"].value_counts()
```

```python
df["Type"].value_counts().plot(kind="bar")

plt.title("Product Type Distribution")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
df["Machine failure"].value_counts()
```

```python
df["Machine failure"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")
plt.title("Machine Failure")
plt.show()
```

```python
plt.hist(df["Air temperature [K]"], bins=30)

plt.title("Air Temperature Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df["Process temperature [K]"], bins=30)

plt.title("Process Temperature")
plt.show()
```

```python
plt.hist(df["Rotational speed [rpm]"], bins=25)

plt.title("Rotational Speed")
plt.show()
```

```python
plt.hist(df["Torque [Nm]"], bins=25)

plt.title("Torque")
plt.show()
```

```python
plt.hist(df["Tool wear [min]"], bins=30)

plt.title("Tool Wear")
plt.show()
```

```python
failure = pd.crosstab(df["Type"], df["Machine failure"])

failure.plot(kind="bar")

plt.title("Machine Failure by Product Type")
plt.show()
```

```python
df.groupby("Type")["Torque [Nm]"].mean().plot(kind="bar")

plt.title("Average Torque")
plt.show()
```

```python
df.groupby("Type")["Tool wear [min]"].mean().plot(kind="bar")

plt.title("Average Tool Wear")
plt.show()
```
